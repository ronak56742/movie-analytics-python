```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("movies.csv")

# ---------- BASIC PREVIEW ----------
print("\nFIRST 5 ROWS\n", df.head())
print("\nDATA SHAPE:", df.shape)

# ---------- CREATE PROFIT ----------
df["profit_million"] = df["revenue_million"] - df["budget_million"]

# ---------- SUMMARY ----------
print("\nSUMMARY STATS\n", df[["rating","duration_min","profit_million"]].describe())

# ---------- MOVIES PER GENRE ----------
genre_counts = df["genre"].value_counts()
print("\nMOVIES PER GENRE\n", genre_counts)

genre_counts.plot(kind="bar", title="Movies Per Genre")
plt.ylabel("Count")
plt.show()

# ---------- AVERAGE RATING PER GENRE ----------
avg_rating_genre = df.groupby("genre")["rating"].mean().sort_values(ascending=False)
print("\nAVERAGE RATING BY GENRE\n", avg_rating_genre)

avg_rating_genre.plot(kind="bar", title="Average Rating by Genre")
plt.ylabel("Rating")
plt.show()

# ---------- TOP 5 PROFIT MOVIES ----------
top_profit = df[["title","profit_million"]].sort_values(by="profit_million", ascending=False).head(5)
print("\nTOP 5 PROFIT MOVIES\n", top_profit)

# ---------- RATING VS REVENUE ----------
plt.scatter(df["rating"], df["revenue_million"])
plt.title("Does Rating Affect Revenue?")
plt.xlabel("Rating")
plt.ylabel("Revenue (Million USD)")
plt.show()

# ---------- DURATION VS RATING ----------
plt.scatter(df["duration_min"], df["rating"])
plt.title("Does Duration Affect Rating?")
plt.xlabel("Duration (Minutes)")
plt.ylabel("Rating")
plt.show()

# ---------- YEARLY TREND ----------
movies_per_year = df["release_year"].value_counts().sort_index()
print("\nMOVIES PER YEAR\n", movies_per_year)

movies_per_year.plot(kind="line", marker="o", title="Movies Released Over Time")
plt.ylabel("Number of Movies")
plt.show()
```
